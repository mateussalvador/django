# Iniciando o primeiro projeto com Django
- Como foi dito anteriormente, é importante garantir que temos o Python 3 instalado corretamente. Depois, temos de criar o ambiente virtual, ativá-lo e instalar o django.
- Para garantir que o django foi instalado corretamente, execute o comando `django-admin --version`.

---

# Conferindo todos os comandos do django-admin
- Usamos o comando `django-admin --help`, que exibirá todos os comandos que ele consegue executar.

---

# Iniciando o projeto
- Usamos o comando `django-admin startproject nome_do_projeto .`

## Por que o `.`?
- Inserimos esse ponto para que ele não crie mais uma subpasta para armazenar os arquivos do projeto. Ou seja, se eu uso o comando `django-admin startproject projeto`, ele vai criar uma pasta chamada "projeto" e, dentro dela, um arquivo `manage.py` e outra pasta também chamada "projeto". 
- Para evitar essas duas pastas de mesmo nome, usamos o ponto. Dessa forma, indicamos ao django que queremos que ele insira os arquivos do Django na primeira pasta "projeto", sem ter de criar uma segunda.

---

# Como abrir o servidor de desenvolvimento do Django?
- Usamos o comando `python manage.py runserver numero_da_porta`
- O número da porta é opcional. Se não for passada, por padrão, abrirá na porta 8000.
- Será exibida a tela do foguetinho.

## Nota
- A tela do foguete só é exibida porque estamos com a configuração `DEBUG=True` e porque não temos nenhuma URL configurada.

---

# Diferença entre `django-admin` e `manage.py`
- Se usarmos `python manage.py --help`, perceberemos que os comandos suportados entre ambos são os mesmos.
- A principal diferença é que o `manage.py` armazena uma referência para o arquivo de configurações do nosso projeto em específico. Ou seja, ele saberá como iniciar nosso servidor ou acessar quaisquer configurações do projeto, enquanto o `django-admin` não saberia. 
- **Sendo assim, o `django-admin` será utilizado para criar o projeto. De resto, usaremos em 99% das vezes o `manage.py`.**
