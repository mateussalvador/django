# O que são arquivos estáticos?
- São arquivos que o servidor entrega "prontos", exatamente como estão salos em disco, sem serem gerados ou modificados para cada requisição.
- O conteúdo deles não é personalizado por usuário ou por dados do banco. O mesmo arquivo é enviado para qualquer pessoa que o solicitar.

## Exemplos
- Imagens
- Arquivos CSS
- Arquivos JS
- Fontes
- Em alguns casos, HTML simples que não muda

## Nota
- Páginas HTML montadas a partir de templates e dados do banco de dados são considerados conteúdo dinâmico, e não arquivos estáticos.

# Sobre o nosso head
- O head criado anteriormente continha todo o CSS na tag `<style>`, ou seja, CSS interno.
- Não precisamos ter um arquivo com CSS interno, visto que deste modo ele não poderia ser reaproveitado e, além disso, tornaria o arquivo muito extenso.

## Como usar arquivos estáticos no Django?
- Criamos uma pasta chamada `static`, também utilizando o padrão de namespace, e lá dentro criamos uma pasta chamada `css` para armazenar os arquivos `css`.

## Como chegamos aos arquivos estáticos?
- No topo do arquivo onde queremos carregar um arquivo estático, ao topo, usamos o comando `{% load static %}`
- Após isso, onde queremos utilizar, mostrar ou chamar o arquivo estático, devemos usar `{% static 'caminho_arquivo_estatico' %}`

## `STATIC_URL`
- No `settings.py` temos uma constante `STATIC_URL`. Lá é indicada a url para os arquivos estáticos. 
- Podemos colocar o que quisermos, é basicamente o que aparecerá no navegador ao inspecionar a página. Como o próprio django controla essa URL, não precisamos nos preocupar com isso.
- Por padrão, deixemos como está.

# Arquivos estáticos não atrelados a APPS
- É comum ter arquivos estáticos não atrelados a apps, assim como os templates.
- Por padrão, o Django só busca arquivos estáticos na pasta `static/` dentro dos apps.

## Como referenciar arquivos estáticos fora de apps?
- Para isso, configuramos no `settings.py` uma lista chamada `STATICFILES_DIRS = []` e, dentro dela, passamos o caminho para pastas onde o django deverá buscar arquivos estáticos.
