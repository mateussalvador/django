# Um pequeno problema
- Ao longo do desenvolvimento, criamos arquivos estáticos em diferentes apps e pastas. 
- Quando for para a produção, o Django deixa de buscar esses arquivos automaticamente. Por isso, teremos de configurar o servidor.
- Para que fazer essa configuração, o mais simples é mover todos esses arquivos para uma só pasta. 

## STATIC_ROOT
- Dentro do `settings.py`, criamos essa constante e passamos o caminho para a pasta que armazenará todos esses arquivos.
- É comum chamarmos ela de `static`. Ela pode estar, por exemplo, na `BASE_DIR` do projeto.

## `collectstatic`
- `python manage.py collectstatic` é o comando para que o django colete todos os arquivos estáticos e os armazene na pasta definida em `STATIC_ROOT`.
- O namespace é uma técnica importante, inclusive, para essa etapa. Ao mover os arquivos, ele moverá as pastas e as armazenará dentro de `static/`, separando os arquivos de cada parte do projeto, sejam apps ou outros.
