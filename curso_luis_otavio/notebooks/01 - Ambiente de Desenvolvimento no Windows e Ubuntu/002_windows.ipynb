# Criando ambiente de desenvolvimento no Windows
- Instalar o Git e marcar todas as caixas no instalador.  Depois, só dar next.
- Instalar o Python e marcar a caixa `Add Python to PATH`, depois ir em Customize installation e marque tudo. Por fim, em `Advanced Options`, marque todas as caixas também. Depois é só dar next e finalizar a instalação.
- Instalar o VSCode (apenas dar next)
- Instalar o Chrome (será mais necessário na fase de testes - apenas dar next também)
