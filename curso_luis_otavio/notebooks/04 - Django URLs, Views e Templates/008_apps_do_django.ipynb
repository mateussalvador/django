# APPS do Django
- Como já vimos, é possível criar rotas no `urls.py` do projeto e, dentro dele, criar views que retornam um `HttpResponse`. Acontece que esse arquivo acabaria ficando muito poluído se tivéssemos a intenção de colocar todas as views (que costumam ficar grandes) dentro de `urls.py`.
- Por isso, o Django permite a criação de apps. Esses apps costumam ser "módulos" do nosso projeto.

---

# Apresentação do Projeto
- O projeto do curso consiste num site de receitas, com fotos, hora, tempo de preparo e diferentes campos, formulários, validação de formulários, além de permitir que os usuários enviem receitas que deverão ser aprovadas posteriormente por um administrador.
- Por isso, vamos criar um app chamado `recipes` (receitas em inglês).

## Como criar um app?
- Usamos o comando `python manage.py startapp nome_do_app`. Isso criará uma nova pasta para o aplicativo com novos arquivos.

## `admin.py`
- Aqui, temos um arquivo chamado `admin.py`, onde registramos as tabelas para que possamos realizar CRUD nessas tabelas na área administrativao.

## `apps.py`
- Contém o nome do app. São apenas configurações do aplicativo e quase nunca serão usadas, a menos que precisemos conferir o nome do app.

## `tests.py`
- É usado para testar o nosso projeto

## `views.py`
- É onde criamos as funções de views vistas anteriormente

## `models.py`
- É onde criamos as tabelas no banco de dados por meio de classes no Python

---

# Observação importante
- Assim que criarmos um app, devemos passá-lo em `INSTALLED_APPS`, em projeto / `settings.py`
