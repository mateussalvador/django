# A função render
- Ela recebe os parâmetros:
    - request
    - template_name (o caminho para o template)
    - context (são dicionários onde passamos variáveis que podem ser usadas dentro do template)
    - content-type (o tipo do conteúdo retornado)
    - status (statuscode retornado pela requisição. Por exemplo, por mais que a requisição seja um sucesso e retorne o recurso, ainda assim podemos retornar um código de erro)
    - using

## Exemplo de uso de `context`
- Imagine que temos a seguinte view:
```python
from django.shortcuts import render

def home(request):
    return render(request, 'recipes/home.html', context= {'name': 'Mateus',})
```

- Agora, no nosso template (arquivo html), podemos passar algo como `<h1>Site de {{name}}</h1>`
