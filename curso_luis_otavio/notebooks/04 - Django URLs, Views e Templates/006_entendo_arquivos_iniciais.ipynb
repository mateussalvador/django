# Arquivos Iniciais do Django
- Aqui, o objetivo é compreender qual o objetivo de cada um dos arquivos criados ao iniciar um projeto Django.

## `__init__.py`
- É criado pelo próprio Python. Vem em branco, mas como já foi visto, é bastante útil. Será mais utilizado em apps.

## `asgi.py` e `wsgi.py` 
- São arquivos utilizados em fase de produção para fazer a ligação entre o Django e um servidor web externo.

## `settings.py`
- Configurações gerais do projeto

## `urls.py`
- É a porta de entrada para a aplicação. A partir do endereço digitado, o django carrega as configurações para aquela url específica.
