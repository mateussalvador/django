# Sobre as págians do nosso projeto
- A página inicial é bem padronizada, porém, a página da receita vai ser diferente uma das outras, visto que ela terá de carregar o nome certo da receita, autor, descrição, etc.
- Por isso, é importante que a url se comporte de forma dinâmica e que eu possa, por exemplo, usar `dominio/recipes/3` para identificar a receita de ID 3, sem que eu tenha de configurar manualmente as urls de cada uma das receitas postadas. 
- **Essas urls serão configuradas no `urls.py` do nosso aplicativo, como já foi visto**
- Para não ter de importar as views uma a uma para a url, é mais fácil importar toda a view com `from . import views`

---

# Criando uma url que aponta para uma receita específica
- Nas urls do app, criamos um caminho `path('recipes/', views.recipe)`

## Como tornar essa URL dinâmica?
- Após a barra da rota, passamos um `<nome_do_parametro>`. Ex.: `path('recipes/<id>/', views.recipe)`
- Ao fazer isso, temos que esse parâmetro será enviado automaticamente para a nossa `view`

### Observação
- **É obrigatório inserir a barra ao final da rota, mesmo após o parâmetro!**

## Observação sobre segurança
- Ao passar esse parâmetro para a view, agora nossa url aceitará qualquer conteúdo após a barra. Isso é um problema, visto que, por exemplo, se quisermos passar o id da receita como parâmetro, ele certamente será um número. Ou seja, devemos aceitar apenas números nesse caso.
- Outros valores aceitos, são: `str`, `int`, `slug`, `uuid`, `path`. Isso recebe o nome de **path converters**

### Como especificar o tipo que quero receber?
- Na url, antes do parâmetro, mas dentro de `<>`, passamos o tipo de dado que será aceito. Por exemplo: `path('recipes/<int:id>/', views.recipe)` 


### Nota
- Da forma que fizemos, só será aceita a url com o parâmetro. Se enviarmos apenas `localhost/recipes/`, teremos um not found.
