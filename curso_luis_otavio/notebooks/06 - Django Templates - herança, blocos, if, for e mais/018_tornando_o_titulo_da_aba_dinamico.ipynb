# Como tornar o título da página em algo dinâmico?
- Ao entrar em uma receita, precisamos que o título da página (na aba do navegador) seja dinâmico. 
- Para isso, temos de assegurar que as tags DOCTYPE e HTML estejam no template base, bem como a tag title e a abertura e o fechamento da tag head. Apenas incluíremos o conteúdo do head, e não sua tag.
- Agora, a única coisa que precisamos é criar um outro bloco dentro de `<title></title>`, que será extendido em outras páginas e alterado dinamicamente.

## Exemplo de template base
```html
<!DOCTYPE html>
<html lang="pt-br">
    <head>
        {% include 'recipes/partials/head.html' %}
        <title>{% block title %}{% endblock title %}Recipes</title>
    </head>
    <body>
    </body>
</html>
```

## Exemplo de template do recipe
```html
{% extends 'recipes/pages/home.html' %}

{% block title %}Nome da Receita | {% endblock title %} <!-- AQUI FOI ALTERADO. VEJA QUE O NOME DA PÁGINA DA RECEITA AGORA FICARÁ 'NOME DA RECEITA | RECIPES' -->

{% block content %}
        <main class="main-content-container">
            <div class="main-content main-content-detail container">
                {% include 'recipes/partials/recipe.html' %}
            </div>
        </main>

{% endblock content %}
```

- O nome da receita ficará como "Nome da Receita" apenas por agora, já que depois colocaremos o nome de cada receita propriamente dita.
