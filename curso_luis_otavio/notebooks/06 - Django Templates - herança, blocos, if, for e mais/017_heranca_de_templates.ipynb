# Herança de templates
- Usada para evitar duplicação de código entre templates
- Consiste em gerar uma página e, nas outras, inserimos apenas o que irá mudar.

## Como fazer?
- Criamos um arquivo que será a nossa base. O que queremos mudar, devemos demarcar como sendo o trecho que iremos mudar.
- Para fazer essa demarcação, usamos `{% block nome_do_bloco %}{% endblock nome_do_bloco %}`.
- Ao fazer isso, demarcamos uma área e, qualquer outra página irá inserir conteúdos novos somente dentro desse bloco.

## Como utilizar?
- No arquivo que queremos que herde o template base, usamos no começo o bloco `{% extends 'caminho/arquivo_base.html' %}`
- No início e no final do bloco "novo" a ser inserido naquela página, também fazemos uma delimitação com `{% block nome_do_bloco %}{% endblock nome_do_bloco %}`

### Nota
- O caminho é passado a partir da pasta templates do APP ou dos TEMPLATE_DIRS listados no `settings`.

## Exemplo

### Template Base
```html
<!-- ESTE É O TEMPLATE BASE -->

<!-- Começo do HTML -->

{% block content %}{% endblock content %} <!--Espaço para indicar onde queremos que outros templates que herdem esse template base insiram conteúdos diferentes -->

<!-- Fim  do HTML -->
```

---

### Template filho

```html
<!-- ESTE É O TEMPLATE QUE **HERDA** DO TEMPLATE BASE -->
{% extends 'caminho_template_base/nome_template.html' %}

{% block content %} 
<!-- Aqui, inserimos o conteúdo novo para este template em específico. -->
{% endblock content %} <!--Espaço para indicar onde queremos que outros templates que herdem esse template base insiram conteúdos diferentes -->

```

---

# Onde usaremos isso no projeto?
- Podemos utilizar para que o template de uma receita em específico herde do template da home

---

# Observação sobre blocos
- Apenas podemos criar blocos dentro de páginas, e não de partials.
